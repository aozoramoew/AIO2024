#10. Deal with missing values - Filling

###Gán giá trị trung bình cho các hàng Revenue mang giá trị null

```python
revenue_mean = data_indexed['Revenue (Millions)'].mean()
print("The mean revenue is: ", revenue_mean)
data_indexed['Revenue (Millions)'].fillna(revenue_mean, inplace=True)
```

#11. apply() functions

###Sử dụng hàm để phân loại phim dựa theo rating Good, Bad, Average

```python
def rating_group (rating):
  if rating >= 7.5:
    return 'Good'
  elif rating >= 6.0:
    return 'Average'
  else :
    return 'Bad'

data['Rating_category'] = data['Rating'].apply(rating_group)

data[['Title','Director','Rating','Rating_category']].head(5)
```
